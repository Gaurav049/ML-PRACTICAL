```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Sample data
x = [2, 4, 6, 8, 10, 14, 10, 6]
y = [18, 20, 24, 22, 26, 26, 22, 18]
t = [0, 0, 0, 1, 1, 1, 0, 1]  # class labels

# Convert to numpy arrays and reshape
x = np.array(x)
y = np.array(y)
t = np.array(t)

# Plotting the original points
plt.scatter(x, y, c=t)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("KNN Data Points")
plt.grid(True)
plt.show()

# Prepare data for training
data = list(zip(x, y))
knn = KNeighborsClassifier(n_neighbors=1)  # using k=1
knn.fit(data, t)

# Predicting for a new point
new_x = 7
new_y = 21
new_point = [(new_x, new_y)]
prediction = knn.predict(new_point)

# Plotting the classified point
plt.scatter(x, y, c=t)
plt.scatter(new_x, new_y, c=t[prediction[0]], s=200, edgecolors='k', label=f"New point, class: {prediction[0]}")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("KNN Classification")
plt.legend()
plt.grid(True)
plt.show()
```
